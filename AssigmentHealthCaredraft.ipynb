```python
import pandas as pd
df = pd.read_csv('C:/Users/Administrator/Downloads/healthcare_data.csv')
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
# Convert date columns to datetime
df['admission_date'] = pd.to_datetime(df['admission_date'])
df['discharge_date'] = pd.to_datetime(df['discharge_date'])
```

```python
# This histogram visualizes the age distribution of patients, helping to identify the most common age groups and overall age demographics in the healthcare dataset.
plt.figure(figsize=(10,6))
plt.hist(df['age'], bins=20, edgecolor='black')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# This histogram shows the distribution of bill amounts, revealing the range and frequency of costs incurred by patients.
plt.figure(figsize=(10,6))
plt.hist(df['bill_amount'], bins=20, edgecolor='black')
plt.title('Bill Amount Distribution')
plt.xlabel('Bill Amount ($)')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# This bar chart displays the count of patients by gender, providing insight into the gender balance in the dataset.
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(8,5))
gender_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# This pie chart illustrates the proportion of patients by gender, offering a quick visual representation of gender distribution percentages.
plt.figure(figsize=(8,8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
# This bar chart shows the number of patients in each department, helping to understand which departments handle the most cases.
dept_counts = df['department'].value_counts()
plt.figure(figsize=(12,6))
dept_counts.plot(kind='bar', color='lightgreen')
plt.title('Department Distribution')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# This bar chart represents the distribution of patient statuses, indicating the outcomes or current states of patients.
status_counts = df['status'].value_counts()
plt.figure(figsize=(8,5))
status_counts.plot(kind='bar', color='orange')
plt.title('Patient Status Distribution')
plt.xlabel('Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# This box plot shows the distribution of bill amounts across different departments, highlighting medians, quartiles, and outliers to compare costs by department.
plt.figure(figsize=(12,8))
departments = df['department'].unique()
data = [df[df['department'] == dept]['bill_amount'] for dept in departments]
plt.boxplot(data, labels=departments)
plt.title('Bill Amount Distribution by Department')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Bill Amount ($)')
plt.show()
```

```python
# This scatter plot explores the relationship between patient age and bill amount, potentially revealing trends or correlations between these variables.
plt.figure(figsize=(10,6))
plt.scatter(df['age'], df['bill_amount'], alpha=0.5, color='purple')
plt.title('Age vs Bill Amount')
plt.xlabel('Age')
plt.ylabel('Bill Amount ($)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# This line plot tracks the number of admissions over time, showing trends in hospital admissions on a monthly basis.
df['admission_month'] = df['admission_date'].dt.to_period('M')
admissions_per_month = df.groupby('admission_month').size()
plt.figure(figsize=(12,6))
admissions_per_month.plot(kind='line', marker='o', color='blue')
plt.title('Number of Admissions Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Admissions')
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)
plt.show()
```

```python
# This bar chart highlights the top 10 most common diagnoses, providing insight into the prevalent health issues in the dataset.
diag_counts = df['diagnosis'].value_counts().head(10)
plt.figure(figsize=(12,6))
diag_counts.plot(kind='bar', color='red')
plt.title('Top 10 Diagnoses')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# This bar chart shows the top 10 most frequently administered treatments, indicating the common interventions used.
treat_counts = df['treatment'].value_counts().head(10)
plt.figure(figsize=(12,6))
treat_counts.plot(kind='bar', color='teal')
plt.title('Top 10 Treatments')
plt.xlabel('Treatment')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# This bar chart displays the average bill amount for each department, helping to compare the cost efficiency or typical expenses across departments.
avg_bill_by_dept = df.groupby('department')['bill_amount'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
avg_bill_by_dept.plot(kind='bar', color='gold')
plt.title('Average Bill Amount by Department')
plt.xlabel('Department')
plt.ylabel('Average Bill Amount ($)')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# This stacked bar chart shows the distribution of genders within each department, allowing comparison of gender representation across different hospital departments.
gender_by_dept = df.groupby(['department', 'gender']).size().unstack()
plt.figure(figsize=(12,6))
gender_by_dept.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'])
plt.title('Gender Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Gender')
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# This bar chart illustrates the average age of patients in each department, revealing if certain departments cater to specific age groups.
avg_age_by_dept = df.groupby('department')['age'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
avg_age_by_dept.plot(kind='bar', color='lightblue')
plt.title('Average Age by Department')
plt.xlabel('Department')
plt.ylabel('Average Age')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# This histogram shows the distribution of length of stay (in days) for patients, providing insight into typical hospital stay durations.
df['length_of_stay'] = (df['discharge_date'] - df['admission_date']).dt.days
plt.figure(figsize=(10,6))
plt.hist(df['length_of_stay'], bins=20, edgecolor='black', color='green')
plt.title('Length of Stay Distribution')
plt.xlabel('Length of Stay (Days)')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# This bar chart compares the average bill amounts across different patient statuses, highlighting cost differences based on outcomes.
avg_bill_by_status = df.groupby('status')['bill_amount'].mean().sort_values(ascending=False)
plt.figure(figsize=(8,5))
avg_bill_by_status.plot(kind='bar', color='purple')
plt.title('Average Bill Amount by Status')
plt.xlabel('Status')
plt.ylabel('Average Bill Amount ($)')
plt.xticks(rotation=0)
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# This histogram compares the distribution of bill amounts across different patient statuses, showing how costs vary by outcome.
plt.figure(figsize=(12,6))
statuses = df['status'].unique()
for status in statuses:
    subset = df[df['status'] == status]['bill_amount']
    plt.hist(subset, bins=20, alpha=0.5, label=status, edgecolor='black')
plt.title('Bill Amount Distribution by Status')
plt.xlabel('Bill Amount ($)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# This scatter plot examines the relationship between length of stay and bill amount, helping to identify if longer stays correlate with higher costs.
plt.figure(figsize=(10,6))
plt.scatter(df['length_of_stay'], df['bill_amount'], alpha=0.5, color='orange')
plt.title('Length of Stay vs Bill Amount')
plt.xlabel('Length of Stay (Days)')
plt.ylabel('Bill Amount ($)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# This pie chart shows the proportion of patients in each department, providing a visual breakdown of departmental patient loads.
plt.figure(figsize=(10,10))
plt.pie(dept_counts, labels=dept_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Department Distribution')
plt.axis('equal')
plt.show()
```

```python
# This box plot displays the age distribution across departments, revealing age variations and potential specialization by age group.
plt.figure(figsize=(12,8))
age_data = [df[df['department'] == dept]['age'] for dept in departments]
plt.boxplot(age_data, labels=departments)
plt.title('Age Distribution by Department')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Age')
plt.show()
```

```python
# This stack plot shows the cumulative admissions over time for the top 3 departments, illustrating how departmental loads change monthly.
top_depts = dept_counts.head(3).index
admissions_by_dept_month = df[df['department'].isin(top_depts)].groupby(['admission_month', 'department']).size().unstack().fillna(0)
plt.figure(figsize=(12,6))
plt.stackplot(admissions_by_dept_month.index.astype(str), admissions_by_dept_month.T, labels=top_depts)
plt.title('Admissions Over Time by Top 3 Departments')
plt.xlabel('Month')
plt.ylabel('Number of Admissions')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# This 3D scatter plot visualizes the relationships between age, bill amount, and length of stay, allowing for multi-dimensional analysis of patient data.
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['age'], df['bill_amount'], df['length_of_stay'], c=df['bill_amount'], cmap='viridis', alpha=0.6)
ax.set_title('3D Scatter: Age, Bill Amount, and Length of Stay')
ax.set_xlabel('Age')
ax.set_ylabel('Bill Amount ($)')
ax.set_zlabel('Length of Stay (Days)')
plt.show()
```
