```python
class MyNumbers:
    def __iter__(self):
        self.num = 1
        return self
    def __next__(self):
        if self.num <=5:
            x= self.num 
            self.num +=1
            return x
        else:
            raise StopIteration
myclass = MyNumbers()
myiter = iter(myclass)
 
for x in myiter:
    print(x)
        
```

```python
class TableOfTwo:

    def __iter__(self):
        self.num = 2   
        return self

    def __next__(self):
        if self.num <= 20:
            val = self.num
            self.num += 2
            return val
        else:
            raise StopIteration


obj = TableOfTwo()

for i in obj:
    print(i)
```

```python
# Generator
def my_generator():
    for i in range(1,6):
        yield i
gen = my_generator()

for value in gen:
    print(value)
    
```

```python
import time

start = time.perf_counter()

def my_generator():
    for i in range(1, 6):
        yield i

gen = my_generator()

for value in gen:
    print(value)

end = time.perf_counter()

print("Execution time:", end - start)
```

```python
# ITERATOR is an object that allow you to traverse through all the element of a  collection(like list or tuple) one at a time.
# key points :
# 1. __iter__() → returns the iterator object __next__() → returns next value
# program above

# GENERATOR is a special type of iterator that is defined using a function and the 'yield' keyword
# key points 
# 1. uses yield instead of return 
# 2. genrates values lazily (one at a time).
# 3. more memory efficient than normal function
# program above

# ITERATOR vs GENERATOR

# Iterator: Requires class wilh __iter__() and __next__().
# Generator: Uses function with yield keyword.
# Iterator: More code, manual implementation.
# Generalor: Simple and concise Implementatlon.
# iterator: slightly more control.
# Generalor: More memory efficient.
```

```python
def calculate(num1, num2, choice):
    if choice == '1':
        return num1 + num2
    elif choice == '2':
        return num1 - num2
    elif choice == '3':
        return num1 * num2
    elif choice == '4':
        return num1 / num2
    else:
        raise ValueError("Invalid operation")


def run_calculator():
    try:
        num1 = float(input("first number: "))
        num2 = float(input("second number: "))
    except ValueError:
        print("Please enter valid numbers.")
        return

    print("Choose operation:\n1 - Add\n2 - Subtract\n3 - Multiply\n4 - Divide")
    choice = input("Enter choice (1/2/3/4): ")

    try:
        result = calculate(num1, num2, choice)
        print("Result:", result)
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    run_calculator()
```

```python
import unittest
from calculator import calculate

class TestCalculate(unittest.TestCase):
    def test_add(self):
        self.assertEqual(calculate(5, 3, '1'), 8)
    
    def test_subtract(self):
        self.assertEqual(calculate(10, 4, '2'), 6)
    
    def test_multiply(self):
        self.assertEqual(calculate(6, 7, '3'), 42)
    
    def test_divide(self):
        self.assertEqual(calculate(20, 4, '4'), 5.0)
    
    def test_divide_zero(self):
        with self.assertRaises(ZeroDivisionError):
            calculate(10, 0, '4')
    
    def test_invalid_operation(self):
        with self.assertRaises(ValueError):
            calculate(5, 3, '5')
    
    def test_string_input(self):
        with self.assertRaises(TypeError):
            calculate("5", 3, '1')

if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False) # for Jupyter Notebook
    # unittest.main() #for files like testcalculator.py
```
