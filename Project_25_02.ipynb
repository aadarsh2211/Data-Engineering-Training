```python
# Cross-tabulate thal and condition
thal_condition_crosstab = pd.crosstab(df['thal'], df['condition'])
print("Cross-tabulation of thal vs condition:")
print(thal_condition_crosstab)

# Convert to percentage
thal_condition_percent = thal_condition_crosstab.div(thal_condition_crosstab.sum(axis=1), axis=0) * 100
print("\nPercentage of condition by thal:")
print(thal_condition_percent)
```

```python
# Plot stacked bar chart
thal_condition_percent.plot(kind='bar')
plt.xlabel('Thalassemia (thal)')
plt.ylabel('Percentage (%)')
plt.title('Stacked Bar Chart of Condition by Thalassemia')
plt.legend(title='Condition', labels=['No Disease (0)', 'Disease (1)'])
plt.xticks(rotation=0)
plt.show()
```

```python
# Multi-Factor Risk Analysis using NumPy
import numpy as np

age_filter = np.array(df['age'] > 50)
chol_filter = np.array(df['chol'] > 240)
bp_filter = np.array(df['trestbps'] > 140)

combined_filter = age_filter & chol_filter & bp_filter
# print( df[combined_filter])
high_risk_patients = df[combined_filter]

print(f"Number of patients meeting all criteria: {len(high_risk_patients)}")
# print(high_risk_patients['condition'])
if len(high_risk_patients) > 0:
    disease_rate = high_risk_patients['condition'].mean() * 100
    print(f"Percentage of these patients having heart disease: {disease_rate:.2f}%")
else:
    print("No patients meet all the criteria.")
```

```python

df['risk_score'] = (df['chol'] / 200) + (df['trestbps'] / 120) + df['oldpeak']

# Classify patients as Low Risk, Medium Risk, High Risk
def classify_risk(score):
    if score < 3:
        return 'Low Risk'
    elif score <= 5:
        return 'Medium Risk'
    else:
        return 'High Risk'

df['risk_category'] = df['risk_score'].apply(classify_risk)

print("Risk score statistics:")
print(df['risk_score'].describe())
print("\nRisk category distribution:")
print(df['risk_category'].value_counts())
```

```python
# Visualize of risk categories
risk_counts = df['risk_category'].value_counts()
risk_counts.plot(kind='bar', color=['green', 'yellow', 'red'])
plt.xlabel('Risk Category')
plt.ylabel('Number of Patients')
plt.title('Distribution of Risk Categories')
plt.xticks(rotation=0)
plt.show()
```

# Key Insights

Based on the analysis of the Heart Disease UCI dataset, here are the answers to the key questions:

## 1. Does cholesterol strongly impact heart disease?
- **Correlation Analysis**: The correlation between cholesterol (chol) and heart disease (condition) is approximately 0.12, indicating a weak positive relationship. Higher cholesterol levels are associated with a slightly higher risk of heart disease, but it's not a strong predictor.
- **Visualization**: The scatter plot of Age vs Cholesterol showed a moderate positive correlation (0.21), but cholesterol alone does not strongly determine disease presence.

## 2. Is male population more vulnerable?
- **Gender Analysis**: Males (sex=1) have a higher average cholesterol (239.6) compared to females (sex=0, 261.3), but the disease rate is higher in males (32.3% of males vs 67.7% of females in the dataset).
- **Percentage Distribution**: The dataset has 32.3% males and 67.7% females, with disease rates of 67.7% for males and 32.3% for females when normalized. Males appear more vulnerable based on the higher disease rate among them.

## 3. Does exercise-induced angina significantly increase risk?
- **Exang Analysis**: Patients with exercise-induced angina (exang=1) have a disease rate of 76.4%, compared to 40.6% for those without (exang=0). This indicates a significant increase in risk.
- **Visualization**: The bar chart shows a clear difference, with exang=1 having much higher disease probability.

## 4. Which feature has the strongest correlation with disease?
- **Correlation Matrix**: From the correlation analysis, the features with the strongest correlations to condition are:
  - cp (chest pain type): 0.43
  - thalach (maximum heart rate): -0.42
  - exang (exercise-induced angina): 0.44
  - oldpeak (ST depression): 0.42
  - thal (thalassemia): 0.52
  - ca (number of major vessels): 0.46
- **Strongest**: Thalassemia (thal) has the strongest correlation (0.52), followed by exercise-induced angina (0.44).
