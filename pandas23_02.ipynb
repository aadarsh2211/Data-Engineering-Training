```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sample data representing monthly website  traffic(in thousand)
months = ['Jan','Feb','Mar','Apr','May','Jun']
traffic = [150,2,180,220,250,310]

#create a line plot
plt.plot(months,traffic)
plt.show()
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun']
traffic = [150,200,180,220,250,210]

#create a line plot
plt.plot(months,traffic)
#add labels
plt.xlabel('Month')
plt.ylabel('Monthlt Traffic')
plt.title('Monthly Website Traffic')
# display the plot
plt.show()
```

```python
#customizing the Plot
plt.plot(months, traffic, marker='^', markersize=10, color='red', linestyle='-.', linewidth=3)

#add label and titles
plt.xlabel('Month')
plt.ylabel('Monthlt Traffic')
plt.title('Monthly Website Traffic')

# display the plot
plt.show()
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun']
product_a_revenue = [45,55,40,70,80,75]
product_b_revenue = [35, 40, 50,55,70,68]

plt.plot(months, product_a_revenue, marker='o', linestyle='-', color='blue' ,label='Product A')
plt.plot(months, product_b_revenue, marker='s', linestyle='--', color='red' ,label='Product B')

#add label and titles
plt.xlabel('Month')
plt.ylabel('Monthlt Revenue (in $1000)')
plt.title('Monthly Revenue comparison')

#display a legend to differentiate between Product A and Product B
plt.legend()

# display the plot
plt.show()
```

```python
overs = np.arange(1,7)
Team_a_score = [10,15,5,24,8,10]
Team_b_score = [3, 28, 5,20,15,6]

plt.plot(overs, Team_a_score, marker='o', linestyle='-', color='blue' ,label='Team A')
plt.plot(overs, Team_b_score, marker='s', linestyle='--', color='red' ,label='Team B')
total_a=np.sum(Team_a_score)
total_b=np.sum(Team_b_score)
#add label and titles
plt.xlabel('Overs')
plt.ylabel('Score')
plt.title('Team A vs Team B score')

#display a legend to differentiate between Product A and Product B
plt.legend()
# teams = ['Team A', 'Team B']
# scores = np.array([total_a, total_b])
# winner = teams[np.argmax(scores)]
# print("Winner is", winner)
# display the plot
plt.show()
print("Winner is", ['Team A', 'Team B'][np.argmax(np.array([total_a, total_b]))])
```

```python
import pandas as pd
df = pd.read_csv(r"C:/Users/Administrator/Downloads/IPL_2018_2020.csv")
df
```

```python
print(df.columns)
df.head()
```

```python
import matplotlib.pyplot as plt

# Number of wins per team
wins = df['winner'].value_counts()
# wins.plot(kind='bar')
plt.bar(wins.index, wins.values)
plt.title('Number of Wins per Team in IPL 2018-2020')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Analysis: Top 10 Player of the Match
pom = df['player_of_match'].value_counts().head(10)
# pom.plot(kind='bar', color='skyblue')
plt.bar(pom.index, pom.values, color=['red', 'orange', 'green', 'blue'])
plt.title('Top 10 Players of the Match in IPL 2018-2020')
plt.xlabel('Player')
plt.ylabel('Number of Awards')
plt.xticks(rotation=90)
plt.show()
```
